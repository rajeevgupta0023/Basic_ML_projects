```python
#dataset containing head size and human brian of different peoples
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= (20.0, 10.0)

# Reading Data
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
# collecting data
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
mean_y= np.mean(Y)
m = len(X)
numer = 0
denom = 0
for i in range(m):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x)**2
slop = numer/denom
coeff = mean_y- slop*mean_x

print (slop, coeff)
```

```python
# now checking how good our model is using R square method
ss_total= 0
ss_recidual=0
for i in range(m):
    y_predicted= coeff+ slop*X[i]
    ss_total += (Y[i]-mean_y)**2
    ss_recidual += (Y[i]-y_predicted)**2
r2 = 1-ss_recidual/ss_total
print(r2)
# value near 1 is good, however in case of human behaviour prediction far as 50% value are also acceptable
```
